## 3,4.  Train-Test splitting of dataset and Train using Bernoulli's Classifier

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
clf = BernoulliNB()
clf.fit(X_train, y_train)
```

## 5. For the following types of objects, create a section, and print a) the total positive log-evidence,  b) the total negative log-evidence, c) probability distribution, d) top 3 features values that contribute most to the positive evidence, e) top 3 feature values that contribute the most to the negative evidence. 

### Calculating most positive and negative objects

```python
# Calculate probability of features X given class y i.e (X given y = 1) and (X given y = -1)
# The result is a 2D array of probabilities w.r.t two classes i.e 1 and -1
X_log_probability = clf.feature_log_prob_
#print(X_log_probability)
#Calculate probability of each class y i.e -1 and 1
y_log_probability = clf.class_log_prior_
#print(y_log_probability)
```

```python
# Calculating evidences for all objects in X_test for further usage
# object_arr contains feature values of objects
positive_evidences_list = []
negative_evidences_list = []
positive_features_list = []
negative_features_list = []

# Calculate log of Probability(y = 1)/ log of Probability(y = -1) and checking its ratio to be > 0
log_pos_neg_ratio = y_log_probability[1] - y_log_probability[0]
if log_pos_neg_ratio >= 0:
    positive_log_evidence += log_pos_neg_ratio
else:
    negative_log_evidence += log_pos_neg_ratio
        
for object_arr in X_test:
    object_arr_length = len(object_arr)
    positive_log_evidence = 0
    negative_log_evidence = 0
    positive_features = []
    negative_features = []
    
    # Calculate log of probabilities of features X given y = 1 and -1 ,adding them to the evidence calculations from above and
    # checking them to be > 0 for a positive evidence else a negative evidence
    for i in range(object_arr_length):
        if object_arr[i] == 1:
            log_pos_neg_ratio = X_log_probability[1,i] - X_log_probability[0,i]
        else:
            log_pos_neg_ratio = np.log(1 - np.exp(X_log_probability[1,i])) - np.log(1 - np.exp(X_log_probability[0,i]))
        if log_pos_neg_ratio >= 0:
            positive_log_evidence += log_pos_neg_ratio
            positive_features.append((i, log_pos_neg_ratio))
        else:
            negative_log_evidence += log_pos_neg_ratio
            negative_features.append((i, log_pos_neg_ratio))
    
    positive_evidences_list.append(positive_log_evidence)
    negative_evidences_list.append(negative_log_evidence)
    positive_features_list.append(positive_features)
    negative_features_list.append(negative_features)
```

```python
# Calculate probability of each feature in X_test w.r.t class y 
# The result will be an array containing probabilities of features w.r.t class 1 and class -1

X_predicted_probability = clf.predict_proba(X_test)
print(X_predicted_probability)
```

### 5 - 1. The most positive object with respect to the probabilities.

```python
most_positive_object_ix = np.argmax(X_predicted_probability[:, 1])
print(" 5 - 1 --> The most positive object is at index", most_positive_object_ix)

print(" 5 - 1a) --> The total positive log-evidence = ", positive_evidences_list[most_positive_object_ix])
print(" 5 - 1b) --> The total negative log-evidence = ", negative_evidences_list[most_positive_object_ix])
print(" 5 - 1c) --> Probability Distribution = ", X_predicted_probability[most_positive_object_ix])

#Top 3 features values that contribute most to the positive evidence
top_3_pos_feature = sorted(positive_features_list[most_positive_object_ix], key=lambda x: x[1], reverse=True)[:3]
print(" 5 - 1d) --> Top 3 features contributing most to positive evidence:")
for item in top_3_pos_feature:
    print("            ", feature_list[item[0]], "    ", item[1])
    
#Top 3 features values that contribute most to the negative evidence
top_3_neg_feature = sorted(negative_features_list[most_positive_object_ix], key=lambda x: x[1])[:3]
print(" 5 - 1e) --> Top 3 features contributing most to negative evidence:")
for item in top_3_neg_feature:
    print("            ", feature_list[item[0]], "    ", item[1])
print("\n")
print("As total number of features contributing to negative evidence = 2, it can be seen from below:")
print("\n")
print("Total features contributing to positive evidence: \n",positive_features_list[most_positive_object_ix])
print("\n")
print("Total features contributing to negative evidence: \n",negative_features_list[most_positive_object_ix])
```
